Pearson's correlation coefficient

Return the Pearson's correlation coefficient for two inputs. Pearson's correlation coefficient *r* takes values between -1 and +1. It measures the strength and direction of the linear relationship, where +1 means very strong, positive linear relationship, -1 very strong, negative linear relationship, and 0 no linear relationship.

$$
r = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^{n} (x_i - \bar{x})^2 \sum_{i=1}^{n} (y_i - \bar{y})^2}}
$$




```python
import math,statistics
from scipy.stats import spearmanr,kendalltau

# 均值
print(statistics.mean(f1))
# 相关系数
print("皮尔逊相关系数:",abs(statistics.correlation(f1,labels)))
statistics.mean(f1)
print("最大值:",max(f1))
print("最小值:",min(f1))
print("极差:",max(f1)-min(f1))
print("和:",sum(f1))
print("均值",statistics.median(f1))

# 标准差
print("标准差",statistics.stdev(f1))
# 协方差
print("协方差",statistics.covariance(f1,labels))
# 方差
print("方差",statistics.variance(f1))
# 


```

```python
from scipy.stats import kendalltau, spearmanr, pearsonr
cor = []
kendal = []
spman = []

for i in range(len(features)-1):
    fi = list(data.iloc[:,i])
    # 计算皮尔逊相关系数
    cor.append({features[i]: pearsonr(fi, labels)[0]})
    # 计算 Kendall's tau 相关系数
    kendal.append({features[i]: list(kendalltau(fi, labels))})    
    # 计算 Spearman 相关系数
    spman.append({features[i]: list(spearmanr(fi, labels))})
    
print(cor)
print(kendal)
print(spman)
```

```python
from sklearn.datasets import make_classification
from sklearn.feature_selection import f_classif,f_regression,mutual_info_classif,mutual_info_regression
X = np.array(data)
y = np.array(labels)
f_statistic, f_p_values = f_classif(X, y)
f_res = []
m_res = []
for i in range(len(features)-1):
    fi = features[i]
    f_res.append({fi:list(f_statistic)[i]})

print(f_res)
# 使用 sorted() 对列表按照字典中的数值部分排序
sorted_f = sorted(f_res, key=lambda x: list(x.values())[0])

# 输出排序后的结果
print(sorted_f)
```
